```python
import chemview
chemview.enable_notebook()
```

```python
from chemview import MolecularViewer
import numpy as np

coordinates = np.array([[0.0, 1.1, 0.1], [1, 0, 0]], 'float32')
radii = np.array([1.0, 0.7])
```

```python
# Let's try the surface


def generate_surface(coordinates, radii, resolution=32):
    from chemview.marchingcubes import marching_cubes
    
    area_min = np.array([-5.0, -5.0, -5.0])
    area_max = np.array([5.0, 5.0, 5.0])
    
    x = np.linspace(area_min[0], area_max[0], resolution)
    y = np.linspace(area_min[1], area_max[1], resolution)
    z = np.linspace(area_min[2], area_max[2], resolution)
    
    xv, yv, zv = np.meshgrid(x, y, z)
    
    # First we calculate the density map or whatever it is
    f = np.zeros((x.shape[0], y.shape[0], y.shape[0]))
    for r, c in zip(radii, coordinates):
        f += r**2 / ((xv-c[0])**2 + (yv-c[1])**2 + (zv-c[2])**2)
    
    spacing = tuple((area_max - area_min)/resolution)
    triangles = marching_cubes(f, 1)
    
    faces = []
    verts = []
    for i, t in enumerate(triangles):
        faces.append([i * 3, i * 3 +1, i * 3 +2])
        verts.extend(t)
    
    faces = np.array(faces)
    verts = np.array(verts)*spacing
    return verts + area_min, faces

vertices, faces = generate_surface(coordinates, radii, 32)
```

```python
from chemview.widget import encode_numpy

MolecularViewer(coordinates = coordinates,
                topology = {'atom_types': ["Ar", 'Ne'],
                            'bonds': []},
                surface = {'vertices': encode_numpy(vertices.astype('float32')),
                           'faces': encode_numpy(faces.astype('int32'))})
```
